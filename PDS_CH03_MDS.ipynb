### Filling Missing Data(if any)


```python
titanic_or = pd.read_csv('https://raw.githubusercontent.com/RajenderDBC/MathDataSci_dataset/refs/heads/main/titanic_original.csv')
titanic_or.head()
```

```python
#missing values
titanic_or.isnull().sum()
```

```python
print(sum(titanic_or['age'].isnull())) # == 264 missing values
```

```python
average_age = titanic_or['age'].mean() # get the average age
#use the fillna method to remove null values


# Fill missing Age values (safe assignment)
titanic_or['age'] = titanic_or['age'].fillna(average_age)

```

```python
print(sum(titanic_or['age'].isnull()))
```

## Average age
####  What is the average age for a female or a male?
#### Pandas has a built-in function for this, called groupby.

```python
titanic_or.groupby('sex')['age'].mean()
```
